# Notebook to inspect a case in an enriched event log

```python
import pandas as pd
from datetime import datetime
```

```python
INPUT_FILE_NAME = 'Xrtfm_full.csv' # the enriched event log
CASE_ID_COL = 'case:concept:name'
TIMESTAMP_COL = 'time:timestamp'
ACTIVITY_COL = 'concept:name'
```

```python
df = pd.read_csv(INPUT_FILE_NAME,  dtype={'org:resource': str, 'matricola': str}, parse_dates=[TIMESTAMP_COL])
print(len(df), 'rows read from', INPUT_FILE_NAME)
# df = df.drop(columns=['lifecycle:transition'])
df = df.sort_values(TIMESTAMP_COL, ignore_index=True)
# df.info()
df.head()
```

```python
# Select the case here
case_id = 'S49055' 
case_df = df[df[CASE_ID_COL] == case_id]
# case_file_name = 'case_'+ case_id + '.csv' # uncomment to export the case to view in external tools
# case_df.to_csv(case_file_name, index=False)
case_df
```

```python
# Select some event attributes here
selected_attributes = ['dismissal', 
                       'amount', 
                       'paymentAmount', 
                       'expense', 
                       'outstanding_balance', 
                       'outstanding_balance_without_expense', 
                       'outstanding_balance_without_penalty']

short_case_df = case_df[[TIMESTAMP_COL, ACTIVITY_COL]+selected_attributes] 
# case_file_name = 'case_'+ case_id + '.csv'
# short_case_df.to_csv(case_file_name, index=False)
short_case_df
```
